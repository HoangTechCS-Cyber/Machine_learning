```python
#sigmoid funcion
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1/(1+ np.exp(-x))

x = np.array([1, -5, -4, 3, -2])

output = sigmoid(x)
print(output)
plt.plot(output)
```
